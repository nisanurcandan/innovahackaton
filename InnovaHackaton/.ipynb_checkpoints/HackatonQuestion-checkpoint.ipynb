# TASK - 1

```python
import pandas as pd

pd.set_option('display.max_columns', None)

df = pd.read_excel('hackathon 2024 - Yapay Zeka.xlsx')

print(df.isnull().sum())

df['TIME_STAMP'] = pd.to_datetime(df['TIME_STAMP'])

df['WEEKDAY'] = df['TIME_STAMP'].dt.weekday
df['HOUR'] = df['TIME_STAMP'].dt.hour
df['MINUTE'] = df['TIME_STAMP'].dt.minute

print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df['TIME_STAMP'], df['DOWNLOAD'], label='Download')
plt.plot(df['TIME_STAMP'], df['UPLOAD'], label='Upload')
plt.xlabel('Time')
plt.ylabel('Traffic')
plt.legend()
plt.title('Traffic Over Time')
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)

df_clean = df.dropna()

X = df_clean[['DOWNLOAD', 'UPLOAD']].to_numpy()

iso_forest = IsolationForest(contamination=0.01)
df_clean = df_clean.copy()
df_clean['ANOMALY'] = iso_forest.fit_predict(X)

anomalies = df_clean[df_clean['ANOMALY'] == -1]

plt.figure(figsize=(12, 6))
plt.plot(df_clean['TIME_STAMP'], df_clean['DOWNLOAD'], label='Download')
plt.plot(df_clean['TIME_STAMP'], df_clean['UPLOAD'], label='Upload')
plt.scatter(anomalies['TIME_STAMP'], anomalies['DOWNLOAD'], color='red', label='Anomalies', marker='x')
plt.xlabel('Time')
plt.ylabel('Traffic')
plt.legend()
plt.title('Anomaly Detection in Traffic Data')
plt.show()

print(anomalies.head())
```

# TASK - 2

```python
df_hourly = df_clean.set_index('TIME_STAMP').resample('H').mean().reset_index()

df_hourly.head()
```

```python
from statsmodels.tsa.arima.model import ARIMA

model_download = ARIMA(df_hourly['DOWNLOAD'], order=(5,1,0))
model_download_fit = model_download.fit()
df_hourly['PREDICTED_DOWNLOAD'] = model_download_fit.predict(start=0, end=len(df_hourly)-1, dynamic=False)

model_upload = ARIMA(df_hourly['UPLOAD'], order=(5,1,0))
model_upload_fit = model_upload.fit()
df_hourly['PREDICTED_UPLOAD'] = model_upload_fit.predict(start=0, end=len(df_hourly)-1, dynamic=False)

df_hourly['ANOMALY_DOWNLOAD'] = (df_hourly['DOWNLOAD'] - df_hourly['PREDICTED_DOWNLOAD']).abs() > 2 * df_hourly['PREDICTED_DOWNLOAD'].std()
df_hourly['ANOMALY_UPLOAD'] = (df_hourly['UPLOAD'] - df_hourly['PREDICTED_UPLOAD']).abs() > 2 * df_hourly['PREDICTED_UPLOAD'].std()

plt.figure(figsize=(12, 6))
plt.plot(df_hourly['TIME_STAMP'], df_hourly['DOWNLOAD'], label='Download')
plt.plot(df_hourly['TIME_STAMP'], df_hourly['PREDICTED_DOWNLOAD'], label='Predicted Download')
plt.scatter(df_hourly[df_hourly['ANOMALY_DOWNLOAD']]['TIME_STAMP'], df_hourly[df_hourly['ANOMALY_DOWNLOAD']]['DOWNLOAD'], color='red', label='Anomalies', marker='x')
plt.xlabel('Time')
plt.ylabel('Download Traffic')
plt.legend()
plt.title('Time Series Forecasting - Download Traffic Anomalies')
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(df_hourly['TIME_STAMP'], df_hourly['UPLOAD'], label='Upload')
plt.plot(df_hourly['TIME_STAMP'], df_hourly['PREDICTED_UPLOAD'], label='Predicted Upload')
plt.scatter(df_hourly[df_hourly['ANOMALY_UPLOAD']]['TIME_STAMP'], df_hourly[df_hourly['ANOMALY_UPLOAD']]['UPLOAD'], color='red', label='Anomalies', marker='x')
plt.xlabel('Time')
plt.ylabel('Upload Traffic')
plt.legend()
plt.title('Time Series Forecasting - Upload Traffic Anomalies')
plt.show()

print(df_hourly[(df_hourly['ANOMALY_DOWNLOAD']) | (df_hourly['ANOMALY_UPLOAD'])].head())
```

# BONUS TASK

```python
df_hourly_traffic = df_clean.set_index('TIME_STAMP').resample('H').sum().reset_index()

df_hourly_traffic.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df_hourly_traffic['TIME_STAMP'], df_hourly_traffic['DOWNLOAD'], label='Download')
plt.plot(df_hourly_traffic['TIME_STAMP'], df_hourly_traffic['UPLOAD'], label='Upload')
plt.xlabel('Time')
plt.ylabel('Traffic')
plt.legend()
plt.title('Hourly Aggregated Traffic Data')
plt.show()
```
