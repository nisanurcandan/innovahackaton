```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('hackathon 2024 - Yapay Zeka.xlsx')
df
```

```python
df.head()
```

```python
model = IsolationForest(n_estimators=100,max_samples='auto',contamination=float(0.2),max_features=1.0)
model.fit(df[['TIME_STAMP']])
```

```python
df['anomalies_time_stamp'] = model.decision_function(df[['TIME_STAMP']])
df['anomaly'] = model.predict(df[['TIME_STAMP']])
df.head(20)
```
